#### Do train and test split with stratify sampling on Outcome variable to maintain the distribution of dependent variable(1 Marks)

```python
### SPLITTING INTO TRAINING AND TEST SETS
from sklearn.model_selection import train_test_split
X = df.drop('Outcome', axis=1)
Y = df[['Outcome']]

test_size = 0.20 # taking 80:20 training and test set
seed = 1  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
print(X_train.shape)
print(X_test.shape)
```

### Train Support Vector Machine Model(1 Marks)

```python
from sklearn.svm import SVC

# Building a Support Vector Machine on train data
SVC = SVC(C=.2, kernel='linear', gamma= 1)
SVC = SVC.fit(X_train, y_train)
# check the accuracy on the training set
print(SVC.score(X_train, y_train))
prediction = SVC.predict(X_test)
```

```python
# your insights
print(SVC.score(X_test, y_test))

# As the test score is higher than the training score the model is overfit.
```

### Scale the data points using MinMaxScaler ( 1 Marks)

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scale = MinMaxScaler()

df['Pregnancies'] = minmax_scale.fit_transform(df[['Pregnancies']])
df['Glucose'] = minmax_scale.fit_transform(df[['Glucose']])
df['BloodPressure'] = minmax_scale.fit_transform(df[['BloodPressure']])
df['SkinThickness'] = minmax_scale.fit_transform(df[['SkinThickness']])
df['Insulin'] = minmax_scale.fit_transform(df[['Insulin']])
df['BMI'] = minmax_scale.fit_transform(df[['BMI']])
df['DiabetesPedigreeFunction'] = minmax_scale.fit_transform(df[['DiabetesPedigreeFunction']])
df['Age'] = minmax_scale.fit_transform(df[['Age']])
df
```

### Try improving the model accuracy using C=1000 (1 Marks)

```python
from sklearn.svm import SVC
SVC = SVC(C=1000, kernel='linear', gamma= 1)
SVC = SVC.fit(X_train, y_train)
# check the accuracy on the training set
print(SVC.score(X_train, y_train))
prediction = SVC.predict(X_test)
print(SVC.score(X_test, y_test))
```
